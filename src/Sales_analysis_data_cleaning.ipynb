```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'D:\Karishma\Data analysis classes\python\datasets\rsw_amazon_sales.csv')
df
```

```python
print(df.shape)
print(df.info())
print(df.isna().sum)
```

```python
#Remove duplicates
duplicate_rows=df.duplicated().sum()
print(duplicate_rows)
```

```python
df=df.drop_duplicates()
```

```python
#filling numerical columns with mediun
num_cols=['Quantity','Unit_Price','Discount','Cost']
for col in num_cols:
    df.loc[:,col] = pd.to_numeric(df[col],errors='coerce')
    df.loc[:,col] = df[col].fillna(df[col].median())
```

```python
#filling categorical columns with mode
cat_cols=['Region','Product_Category','Product_Name']
for col in cat_cols:
    df.loc[:,col]=df[col].fillna(df[col].mode()[0])
```

```python
#date parsing

df.loc[:,'Order_Date']=pd.to_datetime(df['Order_Date'],errors='coerce')
df=df.dropna(subset=['Order_Date'])
```

```python
print("after cleaning", df.shape)
```
